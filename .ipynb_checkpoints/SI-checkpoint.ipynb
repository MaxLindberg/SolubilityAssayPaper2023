# S1

```python
import polars as pl
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
from sklearn.linear_model import LinearRegression
```

```python
data_AB1_40_reps_salt_temp = pl.read_csv('Data/Main/231125 AB(1-40) AisReps BisTemp CisSalt std6to94 with calcs.txt',skip_rows=15,separator=';')
```

## Lower Temp samples (AB1-40)

```python
AB1_40_temp_ini_conces = [30*0.4**x for x in range(7)]*3
AB1_40_temp_ini_conces.sort(reverse=True)
AB1_40_temp_ini_conces
```

```python
AB1_40_temp_neg_ctrl = data_AB1_40_reps_salt_temp.filter(
    pl.col('Group')=='B').filter(
    pl.col('Content').str.contains('Negative'))[' Average over replicates based on user defined formula (340-30/440-40)'].mean()
```

```python
AB1_40_temp = data_AB1_40_reps_salt_temp.filter(
    pl.col('Group')=='B').filter(
    pl.col('Content').str.contains('Sample'))#[' user defined formula: (x  + 216.8 ) / 4088.52   (340-30/440-40)']
AB1_40_temp=AB1_40_temp.with_columns(pl.Series(name='initial conc.',values=AB1_40_temp_ini_conces))
AB1_40_temp
```

```python
plt.plot(AB1_40_temp['initial conc.'],
         AB1_40_temp[' Raw Data (340-30/440-40)'],
         marker='o',linestyle=' ',color='black')
plt.xlabel(r'Initial protein concentration (${\mu}M$)')
plt.ylabel('OPA Fluorescence intensity')
```

```python
colors = plt.get_cmap("Reds",1)
plt.rcParams["font.size"] = 10
plt.figure(figsize=(10,6),facecolor='white')
plt.tight_layout()
models = {}
i=0



df = AB1_40_temp.filter(pl.col('initial conc.')<6)


X = df['initial conc.']
Y = df[' Raw Data (340-30/440-40)']
X=np.array(X).reshape((-1,1))
Y=np.array(Y).reshape((-1,1))
model = LinearRegression().fit(X,Y)
r_sq = model.score(X,Y)
y_pred = model.predict(X)
plt.plot(X,y_pred,color=colors(i))
print('The function: y = ',model.coef_,' * x + ',model.intercept_,' Was fitted with a R^2 of: ',round(r_sq,4))
plt.plot(X,Y,color=colors(i),marker='o',linestyle=' ')
models['temp'] = model
i+=1

plt.grid(which='both')
plt.xlabel('Initial Conc. (µM)')
plt.ylabel('OPA fluorescence')
#plt.legend()
plt.title('Standard curves from linear region, [ini]<6')

#plt.savefig('plots/221012 std curves from linear regions.pdf')
```

```python
key='temp'
AB1_40_temp=AB1_40_temp.with_columns((
    (pl.col(' Raw Data (340-30/440-40)')-models[key].intercept_[0])/models[key].coef_[0][0])
    .alias('conc'))
```

```python
plt.rcParams["figure.dpi"] = 300
plt.rcParams["axes.formatter.use_mathtext"] = True
plt.rcParams["font.family"] = 'serif'
plt.rcParams["font.serif"] = ["cmr10"]
plt.rcParams["mathtext.fontset"]='cm'
plt.rcParams["font.size"] = 12
plt.figure(figsize=(5,3))
plt.plot(AB1_40_temp['initial conc.'],AB1_40_temp['conc'],
         marker='o',linestyle=' ',color='black')
plt.title(r'A$\beta$1-40 solubility at 26$^\circ$C, 20 mM NaP,''\n'' 200$\mu$M EDTA, pH 7.40, 51h')
plt.xlabel(r'Initial Protein Concentration ($\mu$M)')
plt.ylabel(r'Protein Conc. in Solution ($\mu$M)')
plt.tight_layout()

plt.savefig('Plots/S1-AB1-40 26C solplot.pdf')
```

```python
AB1_40_temp.filter(pl.col('initial conc.')>10
                  ).filter(pl.col('initial conc.')<15
                          ).select('conc').write_csv('Data/SI/Processed/AB1-40 temp lowest plateau.csv')
                                                                                                           
```

```python
import polars as pl
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python

df_ABM1_42 = pl.read_csv('Data/SI/filter ads test ABm1-42.txt',separator=';',
                               skip_rows=9)
df_ABM1_42 = df_ABM1_42.drop_nulls()


# "A" = Unfiltered

# "B" = Filtered

groups = pl.Series(['Unfiltered']*21+['Filtered']*21)
conc = [10*0.5**x for x in range(7)]*3
conc.sort(reverse=True)
conc2 = pl.Series(conc*2)
df_ABM1_42.replace('Group',groups)
df_ABM1_42.replace('Content',conc2)
blank_corr_ABM1_42 = df_ABM1_42.select(['Content','Group',' Blank corrected based on Raw Data (360-30/456-20)'])
blank_corr_ABM1_42
```

```python

plt.rcParams["figure.dpi"] = 300
plt.rcParams["axes.formatter.use_mathtext"] = True
plt.rcParams["font.family"] = 'serif'
plt.rcParams["font.serif"] = ["cmr10"]
plt.rcParams["mathtext.fontset"]='cm'

plt.rcParams["font.size"] = 12
plt.figure(figsize=(5,3),facecolor='white')
plt.tight_layout()
colors = ['black','darkred']
models = {}
i=0
for x in ["Unfiltered","Filtered"]:
    df = blank_corr_ABM1_42.filter(pl.col('Group')==x)
    weights = [1/y for y in df.select('Content')]

    X = df.select('Content')
    Y = df.select(' Blank corrected based on Raw Data (360-30/456-20)')/10000
    X=np.array(X).reshape((-1,1))
    Y=np.array(Y).reshape((-1,1))
    plt.plot(X,Y,label=x,color=colors[i],marker='o',linestyle=' ')
    model = LinearRegression().fit(X,Y)#,weights
    r_sq = model.score(X,Y)
    y_pred = model.predict(X)
    plt.plot(X,y_pred,color=colors[i])
    print('The function: y = ',model.coef_,' * x + ',model.intercept_,' Was fitted weigthed with a R^2 of: ',round(r_sq,4),' to ',x)
    plt.annotate('y='+str(round(model.coef_[0][0],3))+'*x + '+str(round(model.intercept_[0],3))+r', R$^2$:'+str(round(r_sq,3)),
                 xy=(0.4,(2-i)/10), xycoords='axes fraction',color=colors[i],fontsize=10)
    
    models[x] = model
    i+=1

plt.xlabel(r'[A$\beta$(M1-42)] ($\mu$M)')
plt.ylabel('OPA FI (a.u.)')
plt.legend()

plt.savefig('Plots/S2-filter adsorption ABm1-42.pdf', bbox_inches="tight")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
AB40_OPA = pd.read_csv('Data/SI/ABm1-40 tube adsorption.txt',
                      skiprows=7,index_col=[0,1,2],sep=';')
AB40_OPA.drop(AB40_OPA.columns[-1],axis=1, inplace=True)
AB40_OPA = AB40_OPA.swaplevel(1,2)
AB40_OPA.index = AB40_OPA.index.set_names(['Group','Content','Well'])
AB40_OPA = AB40_OPA.transpose()
AB40_OPA.index = AB40_OPA.index.set_names(['Time (h)'])
AB40_OPA
```

```python
AB40_blank = AB40_OPA.groupby('Content',axis=1).mean()['Blank B']
```

```python
AB40_blankCorr = AB40_OPA.sub(AB40_blank,axis=0)
```

```python
AB40_norm = pd.DataFrame({'Untreated (Sarstedt)':AB40_blankCorr.loc['0,27']['S']/AB40_blankCorr.loc['0,27']['S'].groupby('Content').mean()[0],
                         'DNA LoBind (Eppendorf)':AB40_blankCorr.loc['0,27']['D']/AB40_blankCorr.loc['0,27']['D'].groupby('Content').mean()[0],
                         'Protein LoBind (Eppendorf)':AB40_blankCorr.loc['0,27']['P']/AB40_blankCorr.loc['0,27']['P'].groupby('Content').mean()[0],
                         'Maxymum Recovery (Axygen)':AB40_blankCorr.loc['0,27']['A']/AB40_blankCorr.loc['0,27']['A'].groupby('Content').mean()[0]})
AB40_norm#.groupby('Content').std()
```

```python
plt.rcParams["figure.dpi"] = 300
plt.rcParams["axes.formatter.use_mathtext"] = True
plt.rcParams["font.family"] = 'serif'
plt.rcParams["font.serif"] = ["cmr10"]
plt.rcParams["mathtext.fontset"]='cm'
plt.rcParams["font.size"] = 12
plt.figure(figsize=(5,3),facecolor='white')
plt.tight_layout()
for tube in AB40_norm.columns:
    plt.errorbar(range(1,5),AB40_norm.groupby('Content').mean()[tube],AB40_norm.groupby('Content').std()[tube],
                marker='o',capsize=2,label=tube)
plt.xticks(range(1,5))
plt.xlabel('# of tubes')
plt.ylabel('Normalized OPA FI (a.u.)')
plt.ylim(0,1.5)
plt.legend(fontsize=10)
plt.savefig('Plots/S3a-AB40-tube-ads.pdf', bbox_inches="tight")
```
